```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
f = "data/raw/pump_sensor_data/20180401_pump_sensor_data.xlsx"
data = pd.read_excel(f,parse_dates=[0],index_col=[0])
```

```python
# --------------------------------------------------------------
# Explore data
# --------------------------------------------------------------

data.head(10)
data.describe()
data.info()
```

```python
# --------------------------------------------------------------
# Transform data
# --------------------------------------------------------------

data = data.drop(["sensor_15"], axis=1)
```

```python
# --------------------------------------------------------------
# Plot data
# --------------------------------------------------------------

data["sensor_00"].plot()
data["sensor_01"].plot()
```

```python
# --------------------------------------------------------------
# Plot data in loop
# --------------------------------------------------------------

for col in data.columns[:5]:
    data[col].plot()
    plt.show()
```

```python
import sys
from glob import glob

path = "data/raw/pump_sensor_data"
files = sorted(glob(path + "/*.xlsx"))

data_combined = pd.concat(
    [
        pd.read_excel(f, parse_dates=[0], index_col=[0]).drop(["sensor_15"], axis=1)
        for f in files
    ]
)

data_combined["sensor_00"].plot()
```
