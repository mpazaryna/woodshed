```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Load the CSV file into a DataFrame
path = '/Users/mpaz/github/ppv-2023/dataset/2023-ppv.csv'
df = pd.read_csv(path)

# Set the style for seaborn plots
sns.set_style("whitegrid")

# Create a figure with two subplots
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# Plot histogram for PPV values
sns.histplot(df["PPV"], ax=ax[0], kde=True, color="skyblue", bins=15)
ax[0].set_title("Histogram of PPV Values")
ax[0].set_xlabel("PPV ($)")
ax[0].set_ylabel("Frequency")

# Plot boxplot for PPV values
sns.boxplot(x=df["PPV"], ax=ax[1], color="lightcoral")
ax[1].set_title("Box Plot of PPV Values")
ax[1].set_xlabel("PPV ($)")

# Adjust the layout
plt.tight_layout()
plt.show()
```

```python
# Load the CSV file into a DataFrame
# df = pd.read_csv("2023-ppv.csv")

import pandas as pd

# Load the CSV file into a DataFrame
path = '/Users/mpaz/github/ppv-2023/dataset/2023-ppv.csv'
df = pd.read_csv(path)

# Adjust pandas display settings to show the entire dataframe
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Display the first few rows of the DataFrame to understand its structure
df.head()
print(df.head())

# Remove commas and dollar signs from the numeric columns and convert them to appropriate data types
df["Visits"] = df["Visits"].str.replace(",", "").astype(int)
df["Net Receipt"] = (
    df["Net Receipt"].str.replace(",", "").str.replace("$", "").astype(float)
)
df["PPV"] = df["PPV"].str.replace(",", "").str.replace("$", "").astype(float)

# Display the cleaned DataFrame
print(df.head())

ppv_stats = df["PPV"].describe()

print(ppv_stats)

# Calculate the Interquartile Range (IQR) for PPV values
Q1 = df["PPV"].quantile(0.25)
Q3 = df["PPV"].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df["PPV"] < lower_bound) | (df["PPV"] > upper_bound)]

outliers[["Group", "Physician", "PPV"]]
print(outliers[["Group", "Physician", "PPV"]])

# Identify the top 5 and bottom 5 podiatrists based on PPV values
top_5 = df.nlargest(5, "PPV")[["Group", "Physician", "PPV"]]
bottom_5 = df.nsmallest(5, "PPV")[["Group", "Physician", "PPV"]]

top_5, bottom_5
print(top_5, bottom_5)
```
