```python
from langchain_community.tools.tavily_search import TavilySearchResults
from langchain_openai import ChatOpenAI
from langchain import hub
from langchain.agents import create_openai_functions_agent
from langchain.agents import AgentExecutor
from langchain.globals import set_verbose, set_debug

def run_langchain_query(verbose=False, debug=False):
    # Set the verbose and debug modes as per the function arguments
    set_verbose(verbose)
    set_debug(debug)

    # Your existing code to set up and run the query
    prompt = hub.pull("hwchase17/openai-functions-agent")
    llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)
    search = TavilySearchResults()
    tools = [search]
    agent = create_openai_functions_agent(llm, tools, prompt)
    agent_executor = AgentExecutor(agent=agent, tools=tools)

    # Execute the query
    return agent_executor.invoke({"input": "what is the weather in San Diego?"})
```

```python
# Show verbose and debug output
response = run_langchain_query(verbose=True, debug=True)
print(response)
```

```python
# Hide verbose and debug output
response = run_langchain_query(verbose=False, debug=False)
print(response)
```
